```python
from wordcloud import WordCloud, STOPWORDS
```

```python
comment_words = ' '
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df.city: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split()
      
#     # Converts each token into lowercase 
#     for i in range(len(tokens)): 
#         tokens[i] = tokens[i].lower() 
    tokens = tokens[1].lower()
          
    comment_words = comment_words + tokens + ' '
  
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.title('Количество предложенией по населенным пунктам', fontsize=20)
plt.show()
```

```python
import seaborn as sns
```

```python
views = df.fillna('kek').groupby(by=['district']).size()
views = pd.DataFrame(data=views)
views
```

```python
temp = df.fillna('kek').groupby(by=['district'])['VIEWS_LAST_WEEK'].sum()
temp = pd.DataFrame(data=temp)
temp
```

```python
for i in range(temp.shape[0]):
    temp.values[i] = temp.values[i] / views.values[i]
temp
```

```python
temp = temp.sort_values(by=['VIEWS_LAST_WEEK'])
```

```python
plt.title("Количество просмотров по районам",fontsize=20)
sns.set(rc={'figure.figsize':(11.7, 8.27)})
ax = sns.barplot(y=temp.index, x=temp.VIEWS_LAST_WEEK)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

df_old = pd.read_csv("realtby.csv", sep=",")
df_new = df.copy()
df_new['agency_name'] = df_old['agency']
df_new = df_new[~df_new['agency_name'].isna()]

watching_agency = df_new.groupby('agency_name')['VIEWS_LAST_WEEK'].mean()
watching_agency = watching_agency.sort_values(ascending=False)

plt.hlines(y=watching_agency.index, xmin=0, xmax=watching_agency, color='r')
plt.plot(watching_agency[0], watching_agency.index[0], "o")
```
