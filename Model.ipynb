```python
import numpy as np 
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from catboost import CatBoostRegressor
```

```python
df = pd.read_csv('realt_prep.csv')
pd.set_option('display.max_columns', 400)
pd.set_option('display.max_rows', 30)
df
```

```python
model = CatBoostRegressor()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df['BUILD_YEAR'] = df['BUILD_YEAR'].str.findall('\d\d\d\d').str[0].astype(int).max()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['PRICE']), df['PRICE'], test_size=0.2, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model, X_train, y_train, cv=10, scoring='r2', n_jobs=-1)
```
